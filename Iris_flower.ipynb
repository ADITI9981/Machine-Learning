```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
iris_df = sns.load_dataset('iris')
```

```python
iris_df
```

```python
sns.scatterplot(data=iris_df, x='sepal_length',y='petal_width');
```

```python
sns.scatterplot(data=iris_df, x='sepal_length',y='petal_length')
```

```python
numeric_cols=['sepal_length','sepal_width','petal_length','petal_width']
```

```python
x=iris_df[numeric_cols]
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3,random_state=42)
```

```python
model.fit(x)
```

```python
model.cluster_centers_
```

```python
x
```

```python
preds = model.predict(x)
```

```python
preds
```

```python
sns.scatterplot(data=x,x='sepal_length',y='petal_length',hue=preds);
centers_x,centers_y = model.cluster_centers_[:,0], model.cluster_centers_[:,2]
plt.plot(centers_x,centers_y,'xb')
```

```python
model.inertia_#lower the inertia better the fit
```

```python
model = KMeans(n_clusters=6,random_state=42)
```

```python
model.fit(x)
```

```python
preds=model.predict(x)
```

```python
preds
```

```python
sns.scatterplot(data=x, x='sepal_length', y='petal_length', hue=preds);
```

```python
model.inertia_
```

```python
options = range(2,11)
inertias = []

for n_clusters in options:
    model = KMeans(n_clusters, random_state=42).fit(x)
    inertias.append(model.inertia_)

plt.title("No. of clusters vs. Inertia")
plt.plot(options, inertias, '-o')
plt.xlabel('No. of clusters (K)')
plt.ylabel('Inertia');
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=1.1,min_samples=4)
```

```python
model.fit(x)
```

```python
model.labels_
```

```python
sns.scatterplot(data=x, x='sepal_length', y='petal_length', hue=model.labels_);
```

```python
 #PCA
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(x)
```

```python
pca.components_
```

```python
transformed = pca.transform(x)
```

```python
transformed
```

```python
sns.scatterplot(x=transformed[:,0],y=transformed[:,1],hue=model.labels_);
```
